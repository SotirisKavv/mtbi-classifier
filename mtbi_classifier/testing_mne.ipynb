```python
import numpy as np
import mne
```

```python
from utils.import_data import import_panda_csv
from utils.constants import SAMPLE_RATE
from utils.node_coordinates import get_coordinates
```

```python
data = np.loadtxt('../data/mTBI/sources_TBI_MEGM001.csv', delimiter=';', skiprows=1)
```

```python
mydata = import_panda_csv('../data/mTBI/sources_TBI_MEGM001.csv')
```

```python
ch_names = [c.strip() for c in import_panda_csv("../data/mTBI/sources_TBI_MEGM001.csv").columns]
coords = get_coordinates(ch_names)
```

```python
info = mne.create_info(ch_names=list(ch_names), sfreq=SAMPLE_RATE, ch_types='mag')
```
